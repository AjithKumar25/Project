```python
import numpy as np
import pandas as pd

print('Imported Successfully !!')
```

```python
df = pd.read_excel('TEN_STATES daily data.xlsx')
testdf1 = pd.read_excel('Ten states Test dates.xlsx')
df
```

```python
df.shape
```

```python
df_combined = pd.concat([df,testdf1])
df_combined
```

```python
df1 = df_combined[['Date','State_Name','Confirmed']]
df1
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1
```

```python
df1['Year'] = pd.DatetimeIndex(df1['Date']).year
df1['Month'] = pd.DatetimeIndex(df1['Date']).month
df1['Day'] = pd.DatetimeIndex(df1['Date']).day
df1['Weekday'] = df1['Date'].dt.day_name()
# df1['Date'] = df1['Date'].dt.strftime('%Y-%m-%d')
df1
```

```python
df1['Index'] = df1['Date'] 
df1 = df1.set_index('Index')
df1
```

```python
from sklearn import preprocessing
df1['Date'] = preprocessing.LabelEncoder().fit_transform(df1['Date'])
df1['State_Name'] = preprocessing.LabelEncoder().fit_transform(df1['State_Name'])
df1['Weekday'] = preprocessing.LabelEncoder().fit_transform(df1['Weekday'])
df1.reset_index(level=0, inplace=True)
df1
```

```python
df1['Index'] = pd.to_datetime(df1['Index'])
testdf = df1[df1['Index']>'2020-11-17']
df1 = df1[df1['Index']<='2020-11-17']
df1
```

```python
testdf
```

```python
X = df1[['Date','Day','Month','Year','Weekday','State_Name']]
Y = df1['Confirmed']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=500)
rf_model = rf.fit(X_train, Y_train)
rf_model
```

```python
val_pred = rf_model.predict(X_val)
val_pred
```

```python
from sklearn import metrics
import math
Avg_Error = math.sqrt(metrics.mean_squared_error(y_true=Y_val,y_pred=val_pred))
print("Average Error on the Validation Data is {}".format(Avg_Error))
```

```python
future_pred = rf_model.predict(testdf[['Date','Day','Month','Year','Weekday','State_Name']])
future_pred
```

```python
results = pd.concat([testdf1,pd.DataFrame(future_pred)],axis = 1)
results = results.rename(columns={0: "Confirmed"})
results.to_excel("Results prediction.xlsx",index=False)
results
```
